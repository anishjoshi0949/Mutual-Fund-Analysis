```
tree = BallTree(coords, metric='haversine')
```

```
radius_in_meters = 50
EARTH_RADIUS = 6371000  # meters
radius_in_radians = radius_in_meters / EARTH_RADIUS
```

```
indices = tree.query_radius(coords, r=radius_in_radians)
```

```
def get_neighbor_ids(i, neighbors):
    return [df.iloc[j].reg_id for j in neighbors if j != i]

df['nearest neighbour'] = [get_neighbor_ids(i, neighbor_ids) for i, neighbor_ids in enumerate(indices)]
```
